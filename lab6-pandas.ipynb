```python
import seaborn as sns
data=sns.load_dataset("planets")
data.head()
```

```python
group=data.groupby("method")#returns an object
group
```

```python
data.shape #shape/size of planet data
```

```python
group.get_group("Radial Velocity")
```

```python
group.get_group("Imaging")
```

```python
for name,size in group:#for finding all name and sizes of groups present in group object return by groupby f/n
    print(name," ",size.shape)
```
